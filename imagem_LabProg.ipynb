## Pixel

Pixel é a menor unidade de informação que compoe uma imagem digital.

Em uma imagem RGB, o pixel é representado por três números de 8 bits associados com as cores R, G e B que correspondem as cores Red, Gren e Blue respectivamente.

Como são números de 8 bits, a faixa de valores é de 0 a 255.

A cor do Pixel será gerada a partir da combinação de cada um dos canais.

Por exemplo: se a combinação de valores for 255, a cor do pixel será branco. Se a combinação for 0, a cor do Pixel será preto.

###Triângulo

A Python Imaging Library usa um sistema de coordenadas de pixel cartesiano, com (0,0) no canto superior esquerdo.

```python
def triangulo(tam):
  WHITE = (255, 255, 255)
  BLACK = (0,0,0)

  nova_imagem = Image.new("RGB", (tam,tam), WHITE)

  for x in range(tam):
    for y in range(tam):
      if x < y:
        nova_imagem.putpixel((x,y), BLACK)
  return nova_imagem
```

```python
t = triangulo(400)
t
```

###Bandeira da Italia

```python
def bandeira_italia(altura):
  GREEN  = (0,146,70)
  WHITE = (241,242,241)
  RED = (206,43,55)

  largura = (3*altura)//2

  nova_imagem = Image.new("RGB", (largura, altura), GREEN)

  faixa = largura//3

  for x in range(faixa):
    for y in range(altura):
      nova_imagem.putpixel((x + faixa, y), WHITE)
      nova_imagem.putpixel((x + 2*faixa, y), RED)
  return nova_imagem
```

```python
bandeira = bandeira_italia(200)
bandeira
```

###Bandeira do Japão

Cálculo do raio da circunferência

raio_circ.png

```python
def bandeira_japao(altura):
  WHITE = (241,242,241)
  RED = (173,35,51)

  largura = (3*altura)//2

  raio = 3*altura//10

  centro = (largura//2, altura//2)

  nova_imagem = Image.new("RGB", (largura, altura), WHITE)

  for x in range(centro[0]-raio, centro[0]+raio):
    for y in range(centro[1]-raio, centro[1]+raio):
      if ( (x-centro[0])**2 + (y-centro[1])**2 <= raio**2):
        nova_imagem.putpixel((x, y), RED)
  return nova_imagem

```

```python
bandeira = bandeira_japao(300)
bandeira
```
